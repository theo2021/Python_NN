```python
import numpy as np
import pickle
from PIL import Image
from neural_network import NN
from matplotlib import pyplot as plt
```

```python
def get_batch_file(file_num):
    with open("cifar-10-batches-py/data_batch_" + str(file_num), 'rb') as file:
        d = pickle.load(file, encoding='bytes')
    return d
def print_image(img_array, save_dir=None):
    img_tr = np.uint8(img_array.reshape((3,32,32)).transpose(1,2,0))
    im = Image.fromarray(img_tr, 'RGB').resize((320,320)).show()
    im.save(save_dir)
def val2onehot(val_array, classes):
    labels = np.zeros((len(val_array), classes))
    for ind,lbl in enumerate(val_array):
        labels[ind,lbl] = 1
    return labels
def validate(classifier, inputs, labels):
    correct=0
    for i,d in enumerate(inputs):
        if np.argmax(labels[i]) == np.argmax(classifier.feed_forward(d)[1]):
            correct+=1
    print('acc', correct/inputs.shape[0])
```

```python
data_1 = get_batch_file(1)
data_2 = get_batch_file(2)
data_3 = get_batch_file(3)
```

```python
classes = 10
labels_1 = val2onehot(data_1[b'labels'], classes)
labels_2 = val2onehot(data_2[b'labels'], classes)
labels_3 = val2onehot(data_3[b'labels'], classes)
inputs_1 = data_1[b'data']
inputs_2 = data_2[b'data']
inputs_3 = data_3[b'data']
mu = np.mean(inputs_1, axis=0)
var = np.sqrt(np.var(inputs_1, axis=0))
inputs_1 = (inputs_1 - mu)/var
inputs_2 = (inputs_2 - mu)/var
inputs_3 = (inputs_3 - mu)/var
```

```python
myNN = NN(inputs_1.shape[1])
myNN.add_layer(10, function='softmax', bias_weight=1)
myNN.compile(loss_function_str='entropy', regularizer={'name': 'ridge', 'args': {'lamb': 0}})
```

```python
tr, val = myNN.train_batch(inputs_1, labels_1, 100, 40, 0.001, 0, True, inputs_2, labels_2)
```

```python
myNN.regularizer.dev(myNN.layers[0].W)
#myNN.layers[0].W
```

```python
plt.plot(tr, color='b', label='training')
plt.plot(val, color='g', label='validation')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend()
```

```python
validate(myNN, inputs_1, labels_1)
validate(myNN, inputs_2, labels_2)
validate(myNN, inputs_3, labels_3)
```

```python
import copy
lsum = 0
step = 0.001
numm = myNN.layers[0].W.shape[0]*myNN.layers[0].W.shape[1]
for i in range(numm):
    #rnd = (np.random.rand(myNN.layers[0].W.shape[0],myNN.layers[0].W.shape[1]))/10000
    rnd = np.zeros(numm)
    rnd[i] = step
    rnd = rnd.reshape(myNN.layers[0].W.shape[0],myNN.layers[0].W.shape[1])
    myNN2 = copy.deepcopy(myNN)
    myNN2.layers[0].W += rnd
    lnn2 = myNN2.loss_function(labels_1[0],myNN2.feed_forward(inputs_1[0])[1])
    #print(lnn, lnn2, (lnn2-lnn))
    rnd = rnd/(step**2)
    lsum+= (lnn2-lnn)*rnd
tmp2 = np.append(lsum.reshape(1,-1), np.zeros(256)).reshape(176,176)
plt.imshow(tmp2)
plt.colorbar()
plt.show()
```
